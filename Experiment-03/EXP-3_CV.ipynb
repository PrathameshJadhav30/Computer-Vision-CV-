# To perform Image Filtering (Smoothing, Sharpening).

```python
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
img = cv2.imread("/content/drive/MyDrive/Computer Vision/Sample.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img
```

#Smoothing

```python
avg_blur = cv2.blur(img, (5,5),0)
avg_blur
```

```python
gaussianBlur = cv2.GaussianBlur(img, (5,5),0)
gaussianBlur
```

```python
cv2.medianBlur= cv2.medianBlur(img, 5)
cv2.medianBlur
```

```python
bilateralBlur = cv2.bilateralFilter(img, 9, 75, 75)
bilateralBlur
```

#Sharpning

```python
sharpn_image = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
```

```python
sharpn = cv2.filter2D(img, -1, sharpn_image)
sharpn
```

```python
fig, ax = plt.subplots(2,3, figsize=(15,10))
ax[0,0].imshow(img)
ax[0,0].set_title("Original")
ax[0,1].imshow(avg_blur)
ax[0,1].set_title("Average Blur")
ax[0,2].imshow(gaussianBlur)
ax[0,2].set_title("Gaussian Blur")
ax[1,0].imshow(cv2.medianBlur)
ax[1,0].set_title("medianBlur")
ax[1,1].imshow(bilateralBlur)
ax[1,1].set_title("bilateralBlur")
ax[1,2].imshow(sharpn)
ax[1,2].set_title("sharpn")
```
