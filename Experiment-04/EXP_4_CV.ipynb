```python
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def edge_detection(image_path='/content/drive/MyDrive/Computer Vision/Sample.jpg',lower_threshold = 100, upper_threshold = 200):
   img = cv2.imread("/content/drive/MyDrive/Computer Vision/Sample.jpg")
   img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python

# Define lower_threshold and upper_threshold here
lower_threshold = 100
upper_threshold = 200

edges = cv2.Canny(img, lower_threshold, upper_threshold)
plt.figure(figsize=(10, 10))
plt.imshow(edges, cmap="gray")
plt.axis('off')
plt.show()
```

```python
edge_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)
edge_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)

# Calculate the magnitude of the edges to combine x and y directions
edges_Sobel = np.sqrt(edge_x**2 + edge_y**2)

# Now you can display the combined edges
plt.imshow(edges_Sobel, cmap="gray")
plt.axis('off')
plt.show()
```

```python
edges_laplacian = cv2.Laplacian(img, cv2.CV_64F)
plt.figure(figsize=(10, 10))
plt.imshow(edges_laplacian, cmap="gray")
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(10, 10))

# Original Image
plt.subplot(2, 2, 1)
plt.imshow(img, cmap="gray")
plt.title("Original Image")
plt.axis('off')

# Canny Edge Detection
plt.subplot(2, 2, 2)
plt.imshow(edges, cmap="gray")
plt.title("Canny Edge Detection")
plt.axis('off')

# Sobel Edge Detection
plt.subplot(2, 2, 3)
plt.imshow(edges_Sobel, cmap="gray")
plt.title("Sobel Edge Detection")
plt.axis('off')

# Laplacian Edge Detection
plt.subplot(2, 2, 4)
plt.imshow(edges_laplacian, cmap="gray")
plt.title("Laplacian Edge Detection")
plt.axis('off')

plt.show()
```
