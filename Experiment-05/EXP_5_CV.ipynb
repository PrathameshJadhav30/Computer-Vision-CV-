```python
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("/content/drive/MyDrive/Computer Vision/Sample.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img
```

```python
h,w=img.shape[:2]
center = (w//2,h//2)
matrix = cv2.getRotationMatrix2D(center=center,angle=45,scale=1)
matrix
rotated = cv2.warpAffine(src=img,M=matrix,dsize=(w,h))
rotated
```

```python
matrix = np.float32([[1, 0, 50], [0, 1, 30]])
shifted = cv2.warpAffine(src=img, M=matrix, dsize=(w, h))
shifted
```

```python
translated_img = cv2.warpAffine(src=img, M=matrix, dsize=(w, h))
translated_img
```

```python
scaled_img = cv2.resize(src=img, dsize=(w*2, h*2), interpolation=cv2.INTER_LINEAR)
scaled_img
```
