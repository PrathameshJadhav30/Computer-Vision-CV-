Q1.Rotate an image by 90 degrees around its center.

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('/content/drive/MyDrive/Sample.jpg')

(h, w) = image.shape[:2]
center = (w // 2, h // 2)

M = cv2.getRotationMatrix2D(center, 90, 1.0)

rotated = cv2.warpAffine(image, M, (w, h))

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()

plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image')
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

Q2.Translate an image to the left by 30 pixels and up by 50 pixels.

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image = cv2.imread('/content/drive/MyDrive/Sample.jpg')

M = np.float32([[1, 0, -30],
                [0, 1, -50]])

(h, w) = image.shape[:2]

# Apply the translation
translated = cv2.warpAffine(image, M, (w, h))

# Display the result
cv2_imshow(image) # Use cv2_imshow instead of cv2.imshow
cv2_imshow(translated) # Use cv2_imshow instead of cv2.imshow
cv2.waitKey(0)
cv2.destroyAllWindows()
```
