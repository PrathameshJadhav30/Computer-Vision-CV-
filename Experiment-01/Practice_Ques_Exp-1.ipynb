Q3. Implement code to resize an image to half its original size.


```python
image = cv2.imread('/content/drive/MyDrive/ISRO.png')

if image is None:
    print("Error: Could not load the image. Please check the file path.")
else:
    original_height, original_width = image.shape[:2]

    new_width = original_width // 2
    new_height = original_height // 2

    resized_image = cv2.resize(image, (new_width, new_height))

    cv2_imshow(image)
    cv2_imshow(resized_image)

```

Q6.  Implement a function to overlay text on an image using OpenCV.

```python
def overlay_text(image_path, text, position=(50, 50), font=cv2.FONT_HERSHEY_SIMPLEX,
                 font_scale=1, color=(255, 255, 255), thickness=2, output_path='output_image.png'):
    """
    Overlays text on an image and saves the result.

    INDIAN SPACE RESEARCH ORGANIZATION (ISRO)
    """
    image = cv2.imread('/content/drive/MyDrive/ISRO.png')

    if image is None:
        print("Error: Could not load the image. Please check the file path.")
        return

    cv2.putText(image, text, position, font, font_scale, color, thickness)

    cv2.imwrite(output_path, image)
    print(f"Image with text saved as {output_path}")

    cv2_imshow(image)

overlay_text(
    image_path='image.png',
    text='ISRO',
    position=(100, 100),
    font_scale=2,
    color=(0, 255, 0),
    thickness=3,
    output_path='output_image.png'
)
```
